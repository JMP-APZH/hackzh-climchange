```python
url_list = ["homog_mo_ALT", "homog_mo_ANT", "homog_mo_BAS", "homog_mo_BER", "homog_mo_CHM", "homog_mo_CHD", 
               "homog_mo_GSB", "homog_mo_DAV", "homog_mo_ELM", "homog_mo_ENG", "homog_mo_GVE", "homog_mo_GRH", "homog_mo_GRC", 
               "homog_mo_CDF", "homog_mo_OTL", "homog_mo_LUG", "homog_mo_LUZ", "homog_mo_MER", "homog_mo_NEU", "homog_mo_PAY", "homog_mo_SBE", 
               "homog_mo_SAM", "homog_mo_SIA", "homog_mo_SIO", "homog_mo_STG", "homog_mo_SAE", "homog_mo_SMA"]
```

```python
for url in url_list:

    url_full = 'https://www.meteoswiss.admin.ch/product/output/climate-data/homogenous-monthly-data-processing/data/' + url + ".txt"
    response = requests.get(url_full)
    path = url_full.replace('https://www.meteoswiss.admin.ch/product/output/climate-data/homogenous-monthly-data-processing/data/', '../data/')

    with open(path, "w") as f:
        f.write(response.text)
    
    df = pd.read_csv(path, encoding="ISO-8859-1", header = 18, delim_whitespace=True)

    convert = path.replace('.txt', '.csv')
    path_save = convert.replace('data', 'data/plotting_data')

    df.to_csv(path_save) 
```

```python
#read the new csv files, calculate mean and concat

for file in url_list:
    path = "../data/plotting_data/" + file + ".csv"
    globals()['df_%s' % file] = pd.read_csv(path)
    
```
