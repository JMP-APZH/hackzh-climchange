```python
df_klimadaten1_son = pd.read_excel('../data/klimadaten1.xlsx', 'Sonnenscheindauer', skiprows=5)
df_klimadaten1_ns = pd.read_excel('../data/klimadaten1.xlsx', 'Jahresniederschlag', skiprows=5)
df_klimadaten1_temp = pd.read_excel('../data/klimadaten1.xlsx', 'Jahrestemperatur', skiprows=5)
df_klimadaten1_snow = pd.read_excel('../data/klimadaten1.xlsx', 'Neuschnee', skiprows=5)

def clean1(df, rows_lst):
    df = df.rename({'Unnamed: 0':'Jahr'}, axis=1)
    df = df.drop(labels=rows_lst, axis=0).reset_index()
    df = df.drop(columns=['index'], axis=0)
    return df

df_klimadaten1_son = clean1(df_klimadaten1_son, [1,2])
df_klimadaten1_ns = clean1(df_klimadaten1_ns, [1,2])
df_klimadaten1_temp = clean1(df_klimadaten1_temp, [1,2])
df_klimadaten1_snow = clean1(df_klimadaten1_snow, [1,2])
df_klimadaten1_son.head()
```

```python
df_klimadaten2_eis = pd.read_excel('../data/klimadaten2.xlsx', 'Eistage', skiprows=5)
df_klimadaten2_frost = pd.read_excel('../data/klimadaten2.xlsx', 'Frosttage', skiprows=5)
df_klimadaten2_som = pd.read_excel('../data/klimadaten2.xlsx', 'Sommertage', skiprows=5)
df_klimadaten2_hitz = pd.read_excel('../data/klimadaten2.xlsx', 'Hitzetage', skiprows=5)
df_klimadaten2_trop = pd.read_excel('../data/klimadaten2.xlsx', 'Tropennächte', skiprows=5)
df_klimadaten2_nied = pd.read_excel('../data/klimadaten2.xlsx', 'Niederschlagstage', skiprows=5)
```

```python
df_klimadaten2_eis = clean1(df_klimadaten2_eis, [0,1,2,3])
df_klimadaten2_frost = clean1(df_klimadaten2_frost, [0,1,2,3])
df_klimadaten2_som = clean1(df_klimadaten2_som, [0,1,2,3])
df_klimadaten2_hitz = clean1(df_klimadaten2_hitz, [0,1,2,3])
df_klimadaten2_trop = clean1(df_klimadaten2_trop, [0,1,2,3])
df_klimadaten2_nied = clean1(df_klimadaten2_nied, [0,1,2,3])
df_klimadaten2_eis.head()
```

```python
df_klimadaten2_eis.columns
```

```python
def clean2(df, string1):
       df = pd.melt(df, id_vars=['Jahr'], value_vars=['Basel-Binningen', 'Bern-Zollikofen 2)', 'Davos',
              'Genf-Cointrin', 'Locarno-Monti', 'Lugano', 'Luzern', 'Neuenburg',
              'Säntis', 'Samedan', 'Sitten', 'St. Gallen', 'Zürich-Fluntern'], value_name=string1)

       df = df.dropna()
       df = df.rename(columns={'variable':'location'})
       return df
```

```python
df_klimadaten2_eis = clean2(df_klimadaten2_eis, 'ice_days')
df_klimadaten2_frost = clean2(df_klimadaten2_frost, 'frost_days')
df_klimadaten2_som = clean2(df_klimadaten2_som, 'summer_days')
df_klimadaten2_hitz = clean2(df_klimadaten2_hitz, 'hot_days')
df_klimadaten2_trop = clean2(df_klimadaten2_trop, 'tropical_night')
df_klimadaten2_nied = clean2(df_klimadaten2_nied, 'precipitation_days')
df_klimadaten2_eis.head()
```

```python
from functools import reduce
dfs = [df_klimadaten2_frost,
        df_klimadaten2_som,
        df_klimadaten2_hitz,
        df_klimadaten2_trop,
        df_klimadaten2_nied]

klima1_final = df_klimadaten2_eis.copy(deep=True)
for df in dfs:
        name = df.columns[-1]
        days = df[name]
        klima1_final = klima1_final.join(days)

klima1_final.info()
```

```python
#co2 emissions

ghg_evolution = pd.read_excel('../data/Evolution_GHG_since_1990_2022-04.xlsx', skiprows=4)
ghg_evolution = ghg_evolution.drop([0,2], axis=0).reset_index()
ghg_evolution = ghg_evolution.drop(columns=['Cat. (2)', 'Basis (4)', 'index'], axis=1)
ghg_evolution = ghg_evolution.rename(columns={'Unnamed: 1': 'category'})
ghg_evolution.iloc[0,0] = 'Total emissions'
ghg_evolution.head(10)
```

```python
ghg_evolution.columns
```

```python
ghg_evolution = pd.melt(ghg_evolution, id_vars=['category'], value_vars=[1990,       1991,       1992,       1993,       1994,
                            1995,       1996,       1997,       1998,       1999,       2000,
                            2001,       2002,       2003,       2004,       2005,       2006,
                            2007,       2008,       2009,       2010,       2011,       2012,
                            2013,       2014,       2015,       2016,       2017,       2018,
                            2019,       2020], value_name='year')

ghg_evolution = ghg_evolution.rename(columns={'variable': 'year', 'year':'emissions'})
ghg_evolution.head()
```

```python
ghg_evolution.to_csv('../data/processed/emissions.csv')
```
