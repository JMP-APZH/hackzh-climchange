```python
from turtle import title


def plot_year_canton(year, canton, canton2):
    year_pick_temp = annual_mean_swiss_temp.loc[(annual_mean_swiss_temp['Year'] >= year) & (annual_mean_swiss_temp['Canton'] == canton)]
    klima_pick = klima_final.loc[(klima_final['Jahr'] >= year) & (klima_final['location'] == canton2)]

    title = "Climate change since born in " + str(year) + " in " + canton
    fig = go.Figure()

    fig.add_trace(go.Scatter(x=year_pick_temp['Year'], y=year_pick_temp['Temperature'],
                        mode='lines+markers',
                        name='Annual Temperature'))
    # fig.add_trace(go.Scatter(x=klima_pick['Jahr'], y=klima_pick['frost_days'],
    #                     mode='lines+markers',
    #                     name='Number of ice days'))
    fig.add_trace(go.Scatter(x=klima_pick['Jahr'], y=klima_pick['hot_days'],
                        mode='lines',
                        name='Number of hot days'))
    fig.add_trace(go.Scatter(x=klima_pick['Jahr'], y=klima_pick['tropical_night'],
                        mode='lines',
                        name='Number of tropical nights'))
    fig.add_trace(go.Scatter(x=co2['Jahr'], y=co2['Treibstoffe \ntotal'],
                        mode='lines',
                        name='co2 Total emission'))

    fig.update_layout(
        title= title,
        width=1200,
        height=700,
        font=dict(
            family="Courier New, monospace",
            size=18,
            color="RebeccaPurple"
        )
    )

    fig.show()
```

```python
plot_year_canton(1994, 'Bern', 'Bern-Zollikofen 2)')
```
